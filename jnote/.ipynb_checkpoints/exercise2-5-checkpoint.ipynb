to the end every time you change the settings, mostly.**

```python
# Param settings
TOTAL_TIME = 100000
ACTION_NUM = 10
INIT_Q = 0
Q_INC_VAR = 0.01
EPS = 0.1
ALPHA = 0.1
```

# Modelling
We geneate $q_*$ values throughout the lifetime offline as below, initially with **equal** values and then take random walks with **normal** increments. This is a nonstationary multi-armed bandit problem model.

```python
import numpy as np
from matplotlib import pyplot as plt

q_all = np.array(np.linspace(INIT_Q,INIT_Q,ACTION_NUM))

for t in range(TOTAL_TIME):
    q_inc = np.random.normal(scale=Q_INC_VAR,size=(1,ACTION_NUM))
    q_all = np.vstack((q_all, q_all[-1]+q_inc))
    
q_best = np.argmax(q_all, axis=1)

plt.plot(np.linspace(0,TOTAL_TIME, TOTAL_TIME+1),q_all)
```

# Learning
We take the learning process, with *sample-average method* or *constant step-size parameter method* as action-value method in parallel, and benchmark the outcomes of **average reward** and **rate of optimal action** for comparison.

```python
import random
import math

Q_sa = np.linspace(0,0,ACTION_NUM)
Q_csp = np.linspace(0,0,ACTION_NUM)
is_explore_sa = map(lambda x: True if x < EPS else False, [random.random() for i in range(TOTAL_TIME)])
is_explore_csp = map(lambda x: True if x < EPS else False, [random.random() for i in range(TOTAL_TIME)])

R_sum_sa = 0
R_log_sa = [0]
hit_sum_sa = 0
prob_log_sa = [0]

R_sum_csp = 0
R_log_csp = [0]
hit_sum_csp = 0
prob_log_csp = [0]

for t in range(1, TOTAL_TIME+1):
    # sample-average method
    A_sa = random.randrange(0, ACTION_NUM) if next(is_explore_sa) else np.argmax(Q_sa)
    R_sa = np.random.normal(loc=q_all[t][A_sa])
    Q_sa[A_sa] += 1/t * (R_sa - Q_sa[A_sa])
    R_sum_sa += R_sa
    R_log_sa.append(R_sum_sa/t)
    hit_sum_sa += 1 if A_sa == q_best[t] else 0
    prob_log_sa.append(hit_sum_sa / t *100)
    # constant step-size parameter method
    A_csp = random.randrange(0, ACTION_NUM) if next(is_explore_csp) else np.argmax(Q_csp)
    R_csp = np.random.normal(loc=q_all[t][A_csp])
    Q_csp[A_csp] += ALPHA * (R_csp - Q_csp[A_csp])
    R_sum_csp += R_csp
    R_log_csp.append(R_sum_csp/t)
    hit_sum_csp += 1 if A_csp == q_best[t] else 0
    prob_log_csp.append(hit_sum_csp / t *100)
    
plt.xlabel("Steps")
plt.ylabel("Average reward")
plt.plot(np.linspace(0,TOTAL_TIME, TOTAL_TIME+1), R_log_sa,label="Sample-average method")
plt.plot(np.linspace(0,TOTAL_TIME, TOTAL_TIME+1), R_log_csp,label="Constant step-size parameter method")
plt.legend()
plt.show()
plt.xlabel("Steps")
plt.ylabel("% Optimal action")
plt.plot(np.linspace(0,TOTAL_TIME, TOTAL_TIME+1), prob_log_sa,label="Sample-average method")
plt.plot(np.linspace(0,TOTAL_TIME, TOTAL_TIME+1), prob_log_csp,label="Constant step-size parameter method")
plt.legend()
```
